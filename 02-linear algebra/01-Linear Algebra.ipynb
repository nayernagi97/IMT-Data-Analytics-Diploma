```python
import numpy as np
```

```python
import matplotlib.pyplot as plt

def plot_vectors(list_v, list_label, list_color):
    _, ax = plt.subplots(figsize=(20, 20))
    ax.tick_params(axis='x', labelsize=14)
    ax.tick_params(axis='y', labelsize=14)
    ax.set_xticks(np.arange(-20, 20))
    ax.set_yticks(np.arange(-20, 20))
    
    
    plt.axis([-20, 20, -20, 20])
    for i, v in enumerate(list_v):
        sgn = 0.4 * np.array([[1] if i==0 else [i] for i in np.sign(v)])
        plt.quiver(v[0], v[1], color=list_color[i], angles='xy', scale_units='xy', scale=1)
        ax.text(v[0]-0.2+sgn[0], v[1]-0.2+sgn[1], list_label[i], fontsize=14, color=list_color[i])

    plt.grid()
    plt.gca().set_aspect("equal")
    plt.show()
```

# Question 1




Solve the following System using `np.linalg.solve()` and using the Elimination Method and determine whether this system is singular or not
$$\begin{cases}
7x_1+5x_2-3x_3=16, \\ 3x_1-5x_2+2x_3=-8, \\ 5x_1+3x_2-7x_3=0, \end{cases}$$

```python
LHS = np.array([[7,5,-3],
                [3,-5,2],
                [5,3,-7]])
RHS=np.array([16,-8,0])
solution = np.linalg.solve(LHS,RHS)
print(solution)
```

```python
LHS = np.array([[7,5,-3],
                [3,-5,2],
                [5,3,-7]])
RHS=np.array([16,-8,0])

full_matrix = np.hstack((LHS,RHS.reshape(3,1)),dtype=np.float16)

full_matrix[0] = full_matrix[0]/full_matrix[0][0]
full_matrix[1] = full_matrix[1]/full_matrix[1][0]
full_matrix[2] = full_matrix[2]/full_matrix[2][0]

full_matrix[1] = full_matrix[1] - full_matrix[0]
full_matrix[2] = full_matrix[2] - full_matrix[0]

# full_matrix[0] = full_matrix[0]/full_matrix[0][0]
full_matrix[1] = full_matrix[1]/full_matrix[1][1]
full_matrix[2] = full_matrix[2]/full_matrix[2][1]


# full_matrix[1] = full_matrix[1] - full_matrix[0]
full_matrix[2] = full_matrix[2] - full_matrix[1]

x3 = full_matrix[2][3]/full_matrix[2][2]
x2 = (full_matrix[1][3]-full_matrix[1][2]*x3)
x1 = (full_matrix[0][3]-full_matrix[0][2]*x3-full_matrix[0][1]*x2)

print(f"x1 = {x1}")
print(f"x2 = {x2}")
print(f"x3 = {x3}")
```

this System is Non Singular
